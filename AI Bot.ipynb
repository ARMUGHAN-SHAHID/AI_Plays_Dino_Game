```python
import win32api as wapi
import time

keyList = []
# for char in "ABCDEFGHIJKLMNOPQRSTUVWXYZ 123456789,.'Â£$/\\":
#     keyList.append(char)
keyList.extend([38,40])

print(keyList) 
def key_check():
    keys = []
    for key in keyList:
        if wapi.GetAsyncKeyState(key):
            if key == 38:
                keys.append("Up")
                print ("Up")
            else:
                keys.append("Down")
                print ("Down")
    return keys
```

```python
def keys_to_output(keys):
#     [Up,Down,Do Nothing]
    output = [0,0,0]
    if 'Up' in keys:
        output[0] = 1
    elif 'Down' in keys:
        output[1] = 1
    else:
        output[2] = 1
    return output


    
```

```python
def main():
    for i in list(range(5))[::-1]:
        print (i+1)
        time.sleep(1)
    
    last_time = time.time()
    
    if os.path.isfile(fileName):
        print ("File exists, loading previous data")
        training_data = list(np.load(fileName))
    else:
        print ("File does not exist, starting fresh")
        training_data = []
    run = True
    while(run):
        oldScreen =  np.array(ImageGrab.grab(bbox=(570, 360, 1330, 499)))
        oldScreen = cv2.cvtColor(oldScreen, cv2.COLOR_BGR2GRAY)
        screen = cv2.resize(oldScreen, (190, 35))
        cv2.imshow('window', screen)
        
        
        newScreen =  cv2.cvtColor(np.array(ImageGrab.grab(bbox=(570, 360, 1330, 499))), cv2.COLOR_BGR2GRAY)

        if (newScreen == oldScreen).all():
            print ("Game Over.")
            run=False
            print (len(training_data))
            np.save(fileName, training_data)
            print ("Saved")
        else:
#             print ("NOT Game Over.")
            keys = key_check()
            output = keys_to_output(keys)
            training_data.append([screen, output])
            
            if len(training_data) % 500 == 0:
                print (len(training_data))
                np.save(fileName, training_data)
                print ("Saved")
        
main()
        
#         cv2.imshow('window',cv2.cvtColor(oldScreen, cv2.COLOR_BGR2GRAY) )
#         if cv2.waitKey(25) & 0xFF == ord('q'):
#             cv2.destroyAllWindows()
#             break
        
    #     print ("Loop took {}".format(time.time()-last_time))
    #     last_time = time.time()
```
